```python
import glob
import os
import pandas as pd
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = "../random"
```

```python
performances = []
for gene_id in range(15,1512):
    slopes = []
    records = glob.glob(data_dir+"/gene_id%d-random_seed*/Random/every_1_iter_num_24_error.csv"%(gene_id))
    for record in records:
        record = pd.read_csv(record)
        fit = scipy.stats.linregress(record.iloc[:,0], record.iloc[:,1])
        slopes.append(fit[0])
    performances.append(np.mean(slopes))
```

```python
n, bins, patches = plt.hist(performances)
plt.show()
```

```python
sum([1 if i <-0.01 else 0 for i in performances])
```
